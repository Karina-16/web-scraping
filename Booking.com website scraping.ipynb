```python
Headers = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'})
```

```python
webpage=requests.get(url, headers=Headers)
```

```python
webpage.content
```

```python
soup=BeautifulSoup(webpage.text)
```

```python
soup
```

```python
Hotel_names= soup.find_all('div', class_="f6431b446c a15b38c233")
```

```python
Hotel_names
```

```python
Hotels=[]

for i in Hotel_names:
    x=(i.text)
    Hotels.append(x)
```

```python
Hotels
```

```python
Rate=soup.find_all("div", class_="a3b8729ab1 d86cee9b25")
```

```python
Rate
```

```python
Ratings=[]

for i in Rate:
    x=(i.text)
    Ratings.append(x)
```

```python
Ratings
```

```python
Rating=[]
for i in Ratings:
    x=(i.split())
    #print(x[1])
    Rating.append(x[1])
```

```python
Rating
```

```python
REviewNO=soup.find_all("div", class_="abf093bdfe f45d8e4c32 d935416c47")
```

```python
REviewNO
```

```python
Review_numbers=[]

for i in REviewNO:
    x=(i.text)
    Review_numbers.append(x)
```

```python
Review_numbers
```

```python
LOC=soup.find_all("span", class_="aee5343fdb def9bc142a")
```

```python
LOC
```

```python
Location=[]

for i in LOC:
    x=(i.text)
    Location.append(x)
```

```python
Location=Location[::2]
```

```python
Location
```

```python
data={"Hotel":Hotels, "Rating":Rating, "Location":Location, "Review_numbers":Review_numbers}
df=pd.DataFrame(data)
df.to_csv("Scrap_data.csv")
```
